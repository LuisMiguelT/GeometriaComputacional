# Uso de polymake: Construcciones de polítopos

## Cubos

Podemos crear un hipercubo de la dimensión señalada:

```polymake
$p= cube(2);
$p->VISUAL;
```

## Polítopos de cruz

```polymake
$p = cross(3);
print ($p->VERTICES);
$p->VISUAL;
```

## Simplex

```polymake
$p = simplex(4);
print $p->VERTICES;
print $p->DIM;
$p->VISUAL;
```

## Pirámides

Podemos construir una pirámide sobre un polítopo dado:

```polymake
$p = cube(2);
$p ->VISUAL;
```

```polymake
$q = pyramid($p);
$q -> VISUAL;
```

# Bipirámides

```polymake
$q = bipyramid($p);
print($q->VERTICES_IN_FACETS);
$q -> VISUAL;
```

## Prisma

```polymake
$p = simplex(2);
$p -> VISUAL;
```

```polymake
$q = prism($p);
$q -> VISUAL;
```

```polymake
$r = prism($q);
$r->VISUAL;
```

```polymake
$p = cyclic(3, 6);
$p -> VISUAL;
```

```polymake
print($p->VERTICES_IN_FACETS);
```
